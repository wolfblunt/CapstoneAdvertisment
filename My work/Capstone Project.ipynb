```python
def data_2011_region(raindata11):
    actual_Rainfall_2011 = raindata11
    
    if actual_Rainfall_2011 < 1128.7600:
        return 'Low Rainfall Region'
    else:
        return 'High Rainfall Region'
    
yearly_rainfall['region_2011'] = yearly_rainfall['2011 Actual Rainfall'].apply(data_2011_region)
```

```python
def data_2013_region(raindata13):
    actual_Rainfall_2013 = raindata13
    
    if actual_Rainfall_2013 < 1277.1120:
        return 'Low Rainfall Region'
    else:
        return 'High Rainfall Region'
    
yearly_rainfall['region_2013'] = yearly_rainfall['2013 Actual Rainfall'].apply(data_2013_region)

yearly_rainfall.head(10)
```
